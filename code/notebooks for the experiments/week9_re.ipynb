```python
# plotting
fig = plt.figure(figsize=(20,10))


plt.plot(ls,nbs_4000_np,'k-',marker = 'o',mfc = 'r', mec = 'r', label = 'number of linear regions, stepsize=0.01')
#plt.fill_between(ls, upp_4000, low_4000, color='red',alpha = 0.1)
plt.legend()

plt.xlabel("L", fontsize = 40)
plt.ylabel("Avg. number of linear regions", fontsize = 40)
#plt.title("Avg. number of linear regions w.r.t. the number of layers L",fontsize=20)


plt.plot(ls,nbs_400000_np,'k-',marker = 'o',mfc = 'b', mec = 'b', label = 'number of linear regions, stepsize=0.0001')
#plt.fill_between(ls, upp_400000, low_400000, color='blue', alpha = 0.1)
plt.legend(prop={'size': 30})

plt.savefig('Report/stepsize.jpg')

plt.show()
```

```python
nbs_L1_upp = nbs_L1 + stds_L1
nbs_L1_low = nbs_L1 - stds_L1

nbs_L10_upp = nbs_L10 + stds_L10
nbs_L10_low = nbs_L10 - stds_L10

nbs_L50_upp = nbs_L50 + stds_L50
nbs_L50_low = nbs_L50 - stds_L50

nbs_L100_upp = nbs_L100 + stds_L100
nbs_L100_low = nbs_L100 - stds_L100
```
